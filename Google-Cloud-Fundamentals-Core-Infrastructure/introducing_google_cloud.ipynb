### Cloud Computing Overwiev
* Başlangıçta `cloud computing` genel bir bakışıyla başlayalım. `Cloud`, bu günlerde sıkça konuşulan bir konu, ancak tam olarak nedir? Amerika Birleşik Devletleri Ulusal Standartlar ve Teknoloji Enstitüsü (NIST), `cloud computing` terimini oluşturdu, ancak bu terimle ilgili özel bir Amerikan bağlamı yok.
* `Cloud computing`, bu beş eşit derecede önemli özelliğe sahip bilgi teknolojilerini (BT) kullanma biçimidir.
  - İlk olarak, müşteriler anlık ve kendi kendine hizmet veren hesaplama kaynaklarına sahiptir. Kullanıcılar, bir web arayüzü aracılığıyla işlem gücü, depolama ve ağa ihtiyaç duymadan insan müdahalesine gerek duymadan alabilirler.
  - İkincisi, müşteriler bu kaynaklara internet üzerinden herhangi bir bağlantıları olduğu her yerden erişebilirler.
  - Üçüncüsü, bulut sağlayıcısı bu kaynakların büyük bir havuzuna sahiptir ve bu kaynakları bu havuzdan kullanıcılara tahsis eder. Bu, sağlayıcının toptan alım yapmasına ve tasarrufları müşterilere geçirmesine olanak tanır. Müşterilerin bu kaynakların kesin fiziksel konumunu bilmeleri veya umursamaları gerekmez.
  - Dördüncüsü, kaynaklar elastiktir, yani esneklerdir, bu nedenle müşteriler de esnek olabilirler. Daha fazla kaynağa ihtiyaçları varsa, daha fazlasını hızlıca alabilirler. Daha azına ihtiyaçları varsa, ölçeklendirebilirler.
  - Ve son olarak, müşteriler yalnızca kullandıkları veya ayırdıkları miktar için öderler. Kaynakları kullanmayı bıraktıklarında ödemeyi de durdururlar. İşte bu, bulutun tanımıdır.

### IaaS and PaaS

* Sanallaştırılmış veri merkezlerine geçiş, müşterilere iki yeni tür sunum sunmuştur. Altyapı olarak hizmet olarak adlandırılan `IaaS (Infrastructure as a Service)` ve platform olarak hizmet veya P`aaS (Platform as a Service)`.

image.png

* `IaaS` sunumları, işlem gücü, depolama ve ağ yeteneklerini sağlar ve fiziksel veri merkezlerine benzer kaynaklar halinde sanal olarak düzenler.

image.png

- Buna karşılık, `PaaS` sunumları, kodu, uygulamanın ihtiyaç duyduğu altyapıya erişim sağlayan kütüphanelere bağlar. Bu, daha fazla kaynağın uygulama mantığına odaklanmasına olanak tanır.

image.png

- IaaS modelinde, müşteriler önceden ayırdıkları kaynaklar için ödeme yaparlar. PaaS modelinde ise müşteriler yalnızca kullandıkları kaynaklar için ödeme yaparlar. Bulut bilişimi evrildikçe, momentum yönetilen altyapı ve yönetilen hizmetlere kaymıştır.

image.png

image.png

- Yönetilen kaynaklar ve hizmetlerden yararlanmak, şirketlerin iş hedeflerine daha fazla odaklanmalarını ve teknik altyapılarını oluşturma ve sürdürme konusunda daha az zaman ve para harcamalarını sağlar.

image.png

- Bu, şirketlerin ürün ve hizmetlerini müşterilere daha hızlı ve güvenilir bir şekilde sunmalarına olanak tanır.

image.png

- Serverless, bulut bilişimin evriminde bir başka adımdır. Altyapı yönetimi gerektirmeyen bir şekilde, geliştiricilerin kodlarına odaklanmalarına izin verir.

image.png

- Google tarafından sunulan serverless teknolojileri, olay tetiklemeli kodu bir ödeme modeli olarak yöneten Cloud Functions'i ve konteyner tabanlı mikroservis uygulamalarını tamamen yönetilen bir ortamda dağıtmalarına izin veren Cloud Run'i içerir.

image.png

image.png

- Bu kursun kapsamı dışında olmasına rağmen, Yazılım olarak Hizmet (SaaS) olarak adlandırılan bir terimi duymuş olabilirsiniz ve bunun ne olduğunu ve bulut ekosistemi içinde nasıl yer aldığını merak etmiş olabilirsiniz. Yazılım olarak Hizmet uygulamaları yerel bilgisayarınıza yüklenmez. Bunun yerine, bulutta bir hizmet olarak çalışır ve doğrudan son kullanıcılar tarafından internet üzerinden tüketilir.

### The Google Cloud Network

* Google Cloud, kendi küresel ağı üzerinde çalışır. Bu türün en büyüğüdür ve Google bunu inşa etmek için yıllarca milyarlarca dolar yatırım yapmıştır. Bu ağ, müşterilerin uygulamaları için mümkün olan en yüksek veri iletim hızını ve en düşük gecikmeyi sağlamak amacıyla dünya genelinde 100'den fazla içerik önbellek düğümünü kullanarak tasarlanmıştır. Bu, yüksek talep olan içeriğin daha hızlı erişim için önbelleğe alındığı yerlerdir, bu da uygulamaların kullanıcı isteklerine en hızlı yanıt süresini sağlayacak konumdan yanıt vermesini sağlar. Google Cloud'un altyapısı beş büyük coğrafi konumda bulunmaktadır: Kuzey Amerika, Güney Amerika, Avrupa, Asya ve Avustralya. Birden fazla konuma sahip olmak, uygulamaları nerede konumlandırmanız gerektiğini seçmenin, kullanılabilirlik, dayanıklılık ve gecikme gibi özellikleri etkilediği için önemlidir; gecikme, bilgi paketinin kaynağından hedefine gitmek için gereken zamanı ölçer. Bu konumların her biri farklı bölgelere ve bölgelere ayrılmıştır. Bölgeler bağımsız coğrafi alanları temsil eder ve bölgelerden oluşur. Örneğin, şu anda üç farklı bölgeyi içeren bir bölge olan Londra veya europe-west2'dir. Bir bölge, Google Cloud kaynaklarının dağıtıldığı bir alandır. Örneğin, Compute Engine'i kullanarak sanal bir makine başlatırsanız, kaynağın yedekliliğini sağlamak için belirttiğiniz bölgede çalışacaktır. Kaynakları farklı bölgelerde çalıştırabilirsiniz. Bu, uygulamaları dünya genelinde kullanıcılara daha yakın getirmek ve bölgelerle ilgili sorunlar, doğal felaket gibi durumlar olduğunda koruma sağlamak için kullanışlıdır. Google Cloud'un bazı hizmetleri, kaynakları çoklu bir bölgeye yerleştirme desteği sunar. Örneğin, Cloud Spanner çoklu bölge yapılandırmaları, veritabanının verilerini yalnızca birden fazla bölgedeki değil, aynı zamanda örnekleme yapılandırmasına göre birden fazla bölgedeki çoklu bölgelere çoğaltmanıza olanak tanır. Bu ek replikalar, Hollanda ve Belçika gibi yapılandırmadaki bölgelere veya iç bölgelere yakın konumlardan düşük gecikme ile veri okumanızı sağlar. Google Cloud şu anda 34 bölgede 103 bölgeyi desteklemektedir, ancak bu sayı sürekli olarak artmaktadır. En güncel sayılara [https://cloud.google.com/about/locations](https://) adresinden ulaşabilirsiniz.

### Environment impact
* Sanal dünya, Google Cloud'un ağını içeren fiziksel altyapı üzerine inşa edilmiştir ve tüm bu hummalı sunucu rafları büyük miktarda enerji tüketir. Mevcut veri merkezleri toplam dünya elektriğinin yaklaşık %2'sini kullanır. Bu bilinçle, Google veri merkezlerinin mümkün olduğunca verimli çalışmasını sağlamak için çalışıyor. Müşterilerimiz gibi, Google da gezegen için doğru şeyleri yapmaya çalışıyor. Google Cloud müşterilerinin kendi çevresel hedefleri olduğunu anlıyoruz ve iş yüklerini Google Cloud üzerinde çalıştırma, bu hedeflere ulaşmanın bir parçası olabilir. Bu nedenle, Google'ın veri merkezleri ISO 14001 sertifikasını alan ilk merkezlerdi, bu da bir organizasyonun kaynak verimliliğini artırarak çevresel performansını geliştirmek için bir çerçeve çizen bir standardı temsil eder. Bu konunun nasıl yapıldığına dair bir örnek olarak, işte Google'ın Finlandiya'daki Hamina'daki veri merkezi. Bu tesis, Google filosundaki en gelişmiş ve verimli veri merkezlerinden biridir. Kullanılan deniz suyu ile soğutma sistemi, enerji tüketimini azaltır ve dünyanın herhangi bir yerindeki ilk türdür. Kuruluş on yılında, Google karbon nötr olma konusunda ilk büyük şirket haline geldi. İkinci on yılımızda ise %100 yenilenebilir enerjiye ulaşan ilk şirket olduk. 2030 yılına kadar tamamen karbonsuz işletilen ilk büyük şirket olmayı hedefliyoruz.

### Security

* Google'ın dokuz hizmetinin her biri milyardan fazla kullanıcıya sahip olduğundan, güvenlik her zaman Google çalışanlarının gündeminde bulunuyor. Güvenlik tasarımı, Google Cloud ve Google hizmetlerinin temelini oluşturan altyapı boyunca yaygın olarak bulunmaktadır.

image.png

* Müşterilerin verilerini güvende tutmak için Google'ın nasıl çalıştığına dair bazı yolları ele alalım. Güvenlik altyapısını, veri merkezlerimizin fiziksel güvenliğinden başlayarak, altyapının altında yatan donanım ve yazılığın nasıl güvence altına alındığına kadar sıralı katmanlar halinde açıklayabiliriz ve son olarak, işletme güvenliğini desteklemek için yerinde bulunan teknik kısıtlamaları ve süreçleri tanımlarız.

image.png

* İlk olarak Donanım altyapısı katmanıyla başlıyoruz ve bu katmanın üç ana güvenlik özelliği bulunmaktadır: İlk olarak, donanım tasarımı ve menşei bulunmaktadır. Google veri merkezlerindeki sunucu kartları ve ağ ekipmanları özel olarak Google tarafından tasarlanmıştır. Google ayrıca sunucu ve çevre birimlerinde kullanılan donanım güvenlik çipi dahil olmak üzere özel çipler tasarlar. Bir sonraki özellik güvenli önyükleme yığınıdır. Google sunucu makineleri, BIOS, önyükleyici, çekirdek ve temel işletim sistemi görüntüsü üzerine kriptografik imzalar gibi çeşitli teknolojiler kullanarak doğru yazılım yığınını başlattıklarından emin olmak için çeşitli teknolojiler kullanır. Bu katmanın son özelliği yer güvenliğidir. Google, kendi veri merkezlerini tasarlar ve inşa eder, bunlar çok katmanlı fiziksel güvenlik önlemlerini içerir. Bu veri merkezlerine sadece çok az sayıda Google çalışanının erişimi vardır. Google ayrıca bazı sunucuları üçüncü taraf veri merkezlerinde barındırır ve burada da veri merkezi operatörünün sağladığı güvenlik katmanlarının üzerine Google tarafından kontrol edilen fiziksel güvenlik önlemleri olduğunu sağlar.

image.png

* Sonraki katman Hizmet dağıtımı katmanıdır ve burada ana özellik, hizmetler arası iletişimin şifrelenmesidir. Google'ın altyapısı, ağ üzerindeki uzak işlem çağrıları ("RPC") verileri için kriptografik gizlilik ve bütünlük sağlar. Google'ın hizmetleri, RPC çağrılarını kullanarak birbirleriyle iletişim kurar. Altyapı, veri merkezleri arasında giden tüm altyapı RPC trafiğini otomatik olarak şifreler. Google, bu varsayılan şifrelemeyi Google veri merkezlerindeki tüm altyapı RPC trafiğine genişletmelerine izin verecek donanım kriptografik hızlandırıcıları dağıtmaya başlamıştır.

image.png

* Daha sonra kullanıcı kimlik katmanı bulunmaktadır. Google'ın merkezi kimlik hizmeti, genellikle son kullanıcılara Google giriş sayfası olarak yansır ve yalnızca bir kullanıcı adı ve şifre sormakla kalmaz. Hizmet, kullanıcıların geçmişte aynı cihazdan veya benzer bir konumdan giriş yapmış olup olmadığı gibi risk faktörlerine dayalı olarak kullanıcılardan ek bilgi isteme yeteneğine sahiptir. Kullanıcılar ayrıca giriş yaparken Universal 2nd Factor (U2F) açık standartına dayalı cihazlar gibi ikincil faktörleri kullanabilirler.

image.png

* Depolama hizmetleri katmanında, dinlenen veri şifrelemesi güvenlik özelliğini buluyoruz. Google'da çoğu uygulama, verilere merkezi olarak yönetilen anahtarlarla uygulanan depolama hizmetlerini dolaylı olarak erişir (diğer bir deyişle, "dosya depolama"). Google ayrıca sabit sürücülerde ve SSD'lerde donanım şifreleme desteği sağlar.

image.png

* Bir üst katman İnternet iletişimi katmanıdır ve bu iki temel güvenlik özelliğini içerir. İnternet üzerinde sunulan Google hizmetleri, kendilerini Google Front End adlı bir altyapı hizmetine kaydederler, bu hizmet tüm TLS bağlantılarının bir genel-özel anahtar çifti ve bir Sertifikalı Otorite (CA) tarafından X.509 sertifikası kullanılarak sonlandırıldığından ve mükemmel ileriye dönük gizliliği destekleyen en iyi uygulamaları takip ettiğinden emin olur. GFE ayrıca Hizmet Dışı Bırakma Saldırılarına karşı korumalar uygular. Ayrıca Hizmet Dışı Bırakma ("DoS") koruması sağlanır. Altyapısının büyüklüğü, Google'ın birçok DoS saldırısını basitçe emebilmesine olanak tanır. Google ayrıca GFE'nin arkasında çalışan bir hizmet üzerinde herhangi bir DoS etkisini daha da azaltan çok katmanlı DoS korumalarına sahiptir.

image.png

* Son katman Google'ın İşletme Güvenliği katmanıdır ve dört temel özelliği sunar. İlk olarak, sızma tespitidir. Kurallar ve makine zekası, Google'ın işletme güvenliği ekiplerine olası olaylar konusunda uyarılar verir. Google, algılama ve yanıt mekanizmalarının etkililiğini ölçmek ve iyileştirmek için Kırmızı Ekip egzersizleri yapar. İkinci olarak, içeriden kaynaklanan riski azaltma vardır. Google, altyapıya yönetici erişimi verilen çalışanların faaliyetlerini sıkı bir şekilde sınırlar ve aktif olarak izler. Sonra çalışanlar için U2F kullanımı bulunmaktadır. Google çalışanlarının Google'a karşı kimlik avı saldırılarına karşı korunması için, çalışan hesapları U2F uyumlu Güvenlik Anahtarlarının kullanımını gerektirir. Son olarak, sıkı yazılım geliştirme uygulamaları vardır. Google merkezi kaynak kontrolü kullanır ve yeni kodun iki tarafça incelenmesini gerektirir. Google ayrıca geliştiricilere belirli sınıf güvenlik hatalarını sokmamalarını sağlayan kütüphaneler sağlar. Ayrıca, Google altyapısındaki veya uygulamalarındaki hataları keşfedebilen ve bize bildiren herkese ödül veren bir Zafiyet Ödül Programı işletir.

### Open source ecosystems

* Bazı organizasyonlar, iş yüklerini buluta taşımaktan çekinirler çünkü belirli bir sağlayıcıya kilitlenme korkusu taşırlar. Ancak, herhangi bir nedenle bir müşteri Google'ın ihtiyaçlarına artık en iyi sağlayıcı olmadığına karar verirse, uygulamalarını başka bir yerde çalıştırma yeteneği sunarız. Google, müşterilere Google dışında diğer seçenekler sunan ekosistemler oluşturmak için açık kaynak lisansları kullanarak teknolojinin temel unsurlarını yayınlamaktadır. Örneğin, Google tarafından geliştirilen makine öğrenimi için açık kaynak yazılım kütüphanesi olan TensorFlow, güçlü bir açık kaynak ekosisteminin merkezindedir. Google, yığının farklı katmanlarında etkileşim yeteneği sağlar. Kubernetes ve Google Kubernetes Engine, müşterilere farklı bulutlarda çalışan mikroservisleri karıştırma ve eşleştirme yeteneği sunar, Google Cloud'un işletme süiti ise müşterilere birden fazla bulut sağlayıcısı üzerinde iş yüklerini izleme imkanı sağlar.
